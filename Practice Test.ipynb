## The spread of a virus is exponential. If 5 people are infected and each person infects 2 other people, how many people have been infected after 5 rounds of infection?

After one round of infection, the initial 5 infected people will each infect 2 other people, resulting in a total of 5 x 2 = 10 new infected people. So the total number of infected people after the first round is 

```python
5 + 10 
```

After the second round of infection, each of the 10 newly infected people will also infect 2 other people, resulting in a total of 10 x 2 = 20 new infected people. So the total number of infected people after the second round is

```python
15 + 20
```

After the third round of infection, each of the 20 newly infected people will infect 2 other people, resulting in a total of 20 x 2 = 40 new infected people. So the total number of infected people after the third round is

```python
35 + 40 
```

After the fourth round of infection, each of the 40 newly infected people will infect 2 other people, resulting in a total of 40 x 2 = 80 new infected people. So the total number of infected people after the fourth round is 

```python
75 + 80
```

After the fifth round of infection, each of the 80 newly infected people will infect 2 other people, resulting in a total of 80 x 2 = 160 new infected people. So the total number of infected people after the fifth round is

```python
155 + 160 
```

**Therefore, after 5 rounds of infection, a total of 315 people have been infected.**
